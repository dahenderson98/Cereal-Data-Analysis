## Exercise 4: America
Quantify the relationship between sugar and ratings. 

Make a plot to visualize this relationship. Superimpose a best fit line (with seaborn) to describe the relationship ([Seaborn RegPlot Documentation]("https://seaborn.pydata.org/generated/seaborn.regplot.html"))

```python
# Make the plot for the data visualization and line of best fit here:
p = sns.regplot(x=df.sugars, y=df.rating, marker="+", color="r")
```


Calculate a statistic describing the relationship between sugar and ratings  (i.e., correlation). Write a statement in plain English interpreting this statistic.

```python
from scipy.spatial.distance import correlation
# Calculate the statistic using this cell:
x_bar = df["sugars"].mean()
x_sd = df["sugars"].std()
y_bar = df["rating"].mean()
y_sd = df["rating"].std()

x_std_val = [(x - x_bar) / x_sd for x in df["sugars"]]
y_std_val = [(y - y_bar) / y_sd for y in df["rating"]]

correlation_coefficient = np.sum( np.matmul(x_std_val,y_std_val) ) / ( len(df.index) - 1 )
correlation_coefficient
```

Write your statement here.

We first find the mean and standard deviation for both our x-variable (sugars) and our y-variable (rating). We then calculate the standard value series for both variables by subtracting the corresponding mean from each value in both data series, and then dividing the result by that variable's standard deviation. We then do element-wise multiplication of both standard value series and sum the resulting series. Finally, we normalize the sum of that series by dividing by our number of data elements minus one. This equation finds the correlation coefficient between our two variables, which is a value ranging between -1 and 1. It describes both the nature and magnitude of the connection between our two variables. 

In this case, cereal's sugars and ratings have a strong negative correlation, with a correlation coefficient of -0.76.

## Exercise 5: America Part 2
Make a few plots comparing the relationships of carbo, sugars, calories, protein, and fat with rating.

```python
# Write your code to compare the various variables with rating below:
multi, ax = plt.subplots( nrows=2, ncols=3, figsize=(28,12))
multi.suptitle("Cereal Rating Trends", fontsize=20)
ax[0][0].scatter(df.carbo,df.rating, color = 'r')
m, b = np.polyfit(df.carbo,df.rating, 1)
ax[0][0].plot(df.carbo, m * df.carbo + b, color="gray")
ax[0][0].title.set_text("Carbs vs. Rating")
ax[0][0].set_xlabel("Carbs")
ax[0][0].set_ylabel("Rating")
ax[0][1].scatter(df.sugars,df.rating, color = 'g')
m, b = np.polyfit(df.sugars,df.rating, 1)
ax[0][1].plot(df.sugars, m * df.sugars + b, color="gray")
ax[0][1].title.set_text("Sugars vs. Rating")
ax[0][1].set_xlabel("Sugars")
ax[0][1].set_ylabel("Rating")
ax[0][2].scatter(df.calories,df.rating, color = 'b')
m, b = np.polyfit(df.calories,df.rating, 1)
ax[0][2].plot(df.calories, m * df.calories + b, color="gray")
ax[0][2].title.set_text("Calories vs. Rating")
ax[0][2].set_xlabel("Calories")
ax[0][2].set_ylabel("Rating")
ax[1][0].scatter(df.protein,df.rating, color = 'orange')
m, b = np.polyfit(df.protein,df.rating, 1)
ax[1][0].plot(df.protein, m * df.protein + b, color="gray")
ax[1][0].title.set_text("Protein vs. Rating")
ax[1][0].set_xlabel("Protein")
ax[1][0].set_ylabel("Rating")
ax[1][1].scatter(df.fat,df.rating, color = 'purple')
m, b = np.polyfit(df.fat,df.rating, 1)
ax[1][1].plot(df.fat, m * df.fat + b, color="gray")
ax[1][1].title.set_text("Fat vs. Rating")
ax[1][1].set_xlabel("Fat")
ax[1][1].set_ylabel("Rating")
multi.delaxes(ax[1][2])
plt.show()
```

Of the variables carbo, sugars, calories, protein, and fat which has the strongest relationship with rating? Justify your answer.

Write your statement here.

Calories has the strongest correlation with rating because it has the steepest regression line for the data, calculated using the numpy polyfit function.
